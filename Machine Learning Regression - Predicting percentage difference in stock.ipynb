```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv('NSE_SCOM_Safaricom.csv')
data
```

```python
# Convert Date column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
```

```python
# Filter for dates between 19th and 31st of December
#2012
dec_dates12 = []

for date in data['Date']:
    if date.month == 12 and date.day >= 20 and date.day <= 30 and date.year == 2012:
        dec_dates12.append(date)

print(dec_dates12)
```

```python
#picking all the december dates and putting them into a dataframe
#concat means concatination in full but in lay mans disecting the data - hence it is picking the whole row
#dt stands for code

december_data = pd.concat([
    data[(data['Date'].dt.month == 12) & (data['Date'].dt.day >= 20) & (data['Date'].dt.day <= 31) & (data['Date'].dt.year == year)]
    for year in range(2012, 2020)
])


december_data
```

```python
december_data['price_difference'] = december_data['Close'] - december_data['Open']

december_data
```

```python
# Train the model
model = LinearRegression()
model.fit(december_data[['Date']], december_data['price_difference'])
```

```python
import numpy as np

dateindicated = np.array([2023-12-20, 2023-12-21, 2023-12-24, 2023-12-27, 2023-12-28]).reshape(-1, 1)
difference = model.predict(dateindicated)
print(difference)
```
