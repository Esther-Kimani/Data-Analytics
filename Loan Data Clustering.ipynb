```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load data
df = pd.read_csv('loan_data.csv')
df.head(10)
```

```python
#StandardScaler() is what scales the data

scaler = StandardScaler()
X = scaler.fit_transform(df.drop([ 'purpose'], axis=1))
X
```

```python
# from the code above, we are dropping Loan_ID,Gender,Married,Education,Self_Employed,Property_Area,Loan_Status
```

```python
# apply KMeans clustering - v the clustering library/means being used is KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
#Now we want to add a new column to our dataframe that has the clusters
# add the cluster labels to the dataframe
#kmeans.labels_ is used to add a label into the dataframe

df['cluster'] = kmeans.labels_
```

```python
#printing the result
# print the first 10 rows of the resulting dataframe
df.head(10)
```

```python
#we rae replacing the data within the columns
cluster_labels = {0: 'Low Risk', 1: 'Medium Risk', 2: 'High Risk'}

df['cluster'] = df['cluster'].replace(cluster_labels) 
df
```
