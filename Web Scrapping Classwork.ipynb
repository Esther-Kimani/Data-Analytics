```python
import requests
from bs4 import BeautifulSoup
```

```python
#CLASS EXERCISE

url = "https://ubunifucollege.com/"

page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
print(soup)
```

```python
pathWays = soup.find_all('span', {"class": "line-height-normal font-weight-600 text-small alt-font margin-5px-bottom some-title text-uppercase display-block"}) 

print(pathWays)
```

```python
courses = []
for course in pathWays:
    courses.append(course.text)
    
print(courses)
```
