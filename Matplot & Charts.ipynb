```python
import matplotlib.pyplot as plt
```

```python
#SIMPLE LINE GRAPH
#Becasue we are using plt we'll use it to plot
# [] is for arrays

x = [474 , 587, 658, 587, 234, 478]
y = ["Jan", "Feb", "Mar", "April", "May", "June"]

plt.plot(y, x)
plt.show()
```

```python
#BAR GRAPH
sales = [474 , 587, 658, 587, 234, 478]
months = ["Jan", "Feb", "Mar", "April", "May", "June"]

plt.bar(months, sales)
plt.show()
```

```python
#PIE CHART
#Purpose of labels is to show what is what in the pie
salesAmount = [474 , 234, 100, 670]
incomeStreams = ["Mpesa", "Fuliza", "Phone calls", "Internet"]

plt.pie(salesAmount, labels = incomeStreams)
plt.show
```

```python
#SCATTER PLOT
towns = ["Nairobi","Kajiado", "Makueni", "Mombasa"]
numberofinfectedpeople = [500, 21, 657, 300]

plt.scatter(towns, numberofinfectedpeople)
plt.show()
```

```python
#GRID GRAPH
#Should have a title for both axis as well as the graph itself
thex = [474 , 587, 658, 587, 234, 478]
they = ["Jan", "Feb", "Mar", "April", "May", "June"]

plt.title("SALES DATA")
plt.xlabel("Months")
plt.ylabel("Sales Volumes(Millions)")
plt.plot(they, thex)
plt.grid()
plt.show()
```

```python
#SIMPLE LINE GRAPH REPRESENTING 2 DATA
#Legend helps us represent the data in a simple line graph
income = [274 , 387, 458, 587, 634, 778]
expense = [200 , 400, 300, 587, 630, 300]
themonths = ["Jan", "Feb", "Mar", "April", "May", "June"]

plt.plot(themonths,income,expense)
plt.legend(["Income","Expense"])
plt.show()
```

```python
#PIE CHART (B)
#Purpose of labels is to show what is what in the pie
#explode is used to pop the data in a pie chart
salesAmountA = [474 , 234, 100, 670]
myexplode = [0.1, 0.05, 0.05, 0.1]
incomeStreamsA = ["Mpesa", "Fuliza", "Phone calls", "Internet"]

plt.pie(salesAmountA, labels = incomeStreamsA, explode = myexplode)
plt.show
```

```python
#PIE CHART (C)
#Purpose of labels is to show what is what in the pie
#explode is used to pop the data in a pie chart
#HEX are specific colour codes

salesAmountB = [474 , 234, 100, 670]
myexplodeB = [0.1, 0.05, 0.05, 0.1]
incomeStreamsB = ["Mpesa", "Fuliza", "Phone calls", "Internet"]
mycolors = ["#43a576", "#557d52", "#47cffc", "#b0e0e6"]

plt.pie(salesAmountB, labels = incomeStreamsB, explode = myexplodeB, colors = mycolors)
plt.show
```

```python
#PIE CHART (D)
#Purpose of labels is to show what is what in the pie
#autopct = '%1.2f%%' means auto percentage
#to move the legend we use loc

salesAmountC = [474 , 234, 100, 670]
incomeStreamsC = ["Mpesa", "Fuliza", "Phone calls", "Internet"]

plt.pie(salesAmountC, labels = incomeStreamsC, autopct = '%1.2f%%')
plt.legend(title = "Income Streams", loc = (0.9,0.8))
plt.show
```

```python
#SCATTER PLOT A
#scatterpoints helps us specify the number of markers
#bbox_to_anchor is used to remove the legend from the graph so as not to interfere with any scatter

townsA = ["Nairobi","Kajiado", "Makueni", "Mombasa"]
hiv = [500, 657, 300, 120]
covid = [50, 67, 800, 12]

plt.scatter(towns, hiv, label ="HIV")
plt.scatter(towns,covid,  label ="COVID")

plt.legend(title = "DISEASES", scatterpoints = 1, fontsize = 8, loc = "upper left", bbox_to_anchor = (1.05,1))
plt.show()
```

```python
#BAR GRAPH A
import numpy as np 

incomeZ = [2000,1200,5000,6000,3200,8000]
expenseZ = [1000,1000,4500,4000,2500,7500]
monthsZ = ["Jan","Feb","Mar","Apr","May","Jun"]

x_axis = np.arange(len(monthsZ))


plt.bar(x_axis -0.2, incomeZ, width=0.3, label = 'Income')
plt.bar(x_axis +0.2, expenseZ, width=0.4, label = 'Expense')

plt.xticks(x_axis, monthsZ)

plt.legend()
plt.show()
```
