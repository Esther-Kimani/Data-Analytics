```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("bank data.csv")
df.head(30)
```

```python
#on a new cell drop all columns with blank data
#this will identify the blank raw and remover it we don't have to specify
df = df.dropna(subset=[])

df.head(10)
```

```python
#then on the dependants we should have pure integers but we have some data that has "3+". So we are going to drop the +

df['Dependents'] = df['Dependents'].replace('3+', 3)
df['LoanAmount'] = df['LoanAmount'].replace('NaN', 0)
df.head(30)
```
