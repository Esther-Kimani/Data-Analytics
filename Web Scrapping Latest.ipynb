```python
#import requests is a library for web scrapping as well

import requests
from bs4 import BeautifulSoup
```

```python
#page = requests.get(url) getting data from that url
#soup = BeautifulSoup(page.content, "html.parser") converting into a html document

url = "https://ubunifucollege.com/software-development"

page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
#print(soup)
```

```python
#printing anything that has been wrapped by 'p' - courseList = soup.find_all('p')
#text-medium-gray text-extra-small text-uppercase is the class name
courseList = soup.find_all('p', {"class": "text-medium-gray text-extra-small text-uppercase"}) 

print(courseList)
```

```python
#Actual web scapping

durations = []
for duration in courseList:
    durations.append(duration.text)
    
print(durations)
```
