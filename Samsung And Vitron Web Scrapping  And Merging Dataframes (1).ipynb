```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.jumia.co.ke/smart-tvs-2282/samsung/#catalog-listing'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

# Find all the elements with the class "item"
items = soup.find_all(class_='name')
item_prices = soup.find_all(class_='prc')


allitems = []
for item in items:
    allitems.append(item.text)
# print(allitems)

itemlist = list(allitems)


prices = []
for price in item_prices:
    prices.append(price.text)
# print(prices)

pricelist = list(prices)


myitem = list(zip(itemlist, pricelist))
myitem
```

```python
import pandas as pd

#converting the above data into a data frame
df = pd.DataFrame({'Brand Samsung': itemlist, 'Price Samsung': pricelist})
df

```

```python
import requests
from bs4 import BeautifulSoup

urlVitron = 'https://www.jumia.co.ke/catalog/?q=vitron+tv'

responseVitron = requests.get(urlVitron)

soupVitron = BeautifulSoup(responseVitron.content, 'html.parser')

# Find all the elements with the class "item"
itemVitron = soupVitron.find_all(class_='name')
vitronPricess = soupVitron.find_all(class_='prc')


vitronTvs = []
for vitronTv in itemVitron:
    vitronTvs.append(vitronTv.text)
#print(vitronTvs)

itemlistvitron = list(vitronTvs)


vitronPrices = []
for vitronPrice in vitronPricess:
    vitronPrices.append(vitronPrice.text)
#print(vitronPrices)

pricelistvitron = list(vitronPrices)


myitemVitron = list(zip(vitronTvs, vitronPrices))
myitemVitron
```
